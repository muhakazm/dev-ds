`BeautifulSoup`

```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL of the Wikipedia page
url = "https://id.wikipedia.org/wiki/Zin%C3%A9dine_Zidane"

# Fetch the page content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the infobox table
infobox = soup.find('table', class_='infobox vcard')

# Print the infobox content
if infobox:
    # print(infobox.prettify())
    print("Infobox found!")
else:
    print("Infobox not found.")
```

```python
# Extract the name
name = infobox.find('caption', class_='fn').get_text()

# Extract the birth date
birth_date = infobox.find('th', string="Tanggal lahir").find_next('td').get_text(strip=True)

# Extract the birthplace
birthplace = infobox.find('th', string="Tempat lahir").find_next('td').get_text(strip=True)

# Extract the image URL
image_tag = infobox.find('img')
image_url = "https:" + image_tag['src'] if image_tag else "Image not found"

# Print the details
print(f"Name: {name}")
print(f"Birth Date: {birth_date}")
print(f"Birthplace: {birthplace}")
print(f"Image URL: {image_url}")
```
