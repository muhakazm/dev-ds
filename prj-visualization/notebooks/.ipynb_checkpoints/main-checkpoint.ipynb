# Widely Used Chart Types for Data Visualization

## **Basic Charts**
1. **Line Chart**: To show trends over time or continuous data.
1. **Bar Chart**: To compare categories or groups.
1. **Histogram**: To show the distribution of a single variable.
1. **Pie Chart**: To show proportions within a whole.
1. **Scatter Plot**: To show relationships between two variables.

## **Advanced Charts**
1. **Box Plot**: To show the distribution and identify outliers.
1. **Heatmap**: To visualize data density or intensity across a two-dimensional space.
1. **Area Chart**: To represent cumulative values over time.
1. **Bubble Chart**: To compare three variables using position and size.
1. **Stacked Bar Chart**: To show part-to-whole relationships for multiple categories.

## **Specialized Charts**
1. **Radar Chart**: To compare multiple variables for a single category.
1. **Tree Map**: To visualize hierarchical data using nested rectangles.
1. **Waterfall Chart**: To show cumulative effects of sequential positive and negative values.
1. **Sunburst Chart**: To display hierarchical data in a circular form.
1. **Chord Diagram**: To show relationships or flows between different categories.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Line Chart - matplotlib

```python
# Example data
years = [2010, 2011, 2012, 2013, 2014, 2015]
sales = [250, 310, 410, 380, 470, 450]

# Create the line chart
plt.figure(figsize=(8, 5))  # Optional: Set the figure size
plt.plot(years, sales, marker='o', linestyle='-', color='b', label='Sales')  # Line chart

# Add chart elements
plt.title("Yearly Sales Growth")  # Add title
plt.xlabel("Year")  # X-axis label
plt.ylabel("Sales (in USD)")  # Y-axis label
plt.grid(True, linestyle='--', alpha=0.7)  # Optional: Add gridlines
plt.legend()  # Add legend
plt.tight_layout()  # Adjust layout to fit everything nicely

# Show the chart
plt.show()
```

Line Chart - seaborn

```python
# Example data
data = {
    "Year": [2010, 2011, 2012, 2013, 2014, 2015],
    "Sales": [250, 310, 410, 380, 470, 450]
}
df = pd.DataFrame(data)

# Create the line chart using seaborn
sns.set(style="whitegrid")  # Optional: Set the style
plt.figure(figsize=(8, 5))  # Optional: Set the figure size
sns.lineplot(data=df, x="Year", y="Sales", marker="o", color="b", label="Sales")

# Add chart elements
plt.title("Yearly Sales Growth")  # Add title
plt.xlabel("Year")  # X-axis label
plt.ylabel("Sales (in USD)")  # Y-axis label
plt.legend()  # Add legend

# Show the chart
plt.show()
```

Line Chart - plotly

```python
# Example data
data = {
    "Year": [2010, 2011, 2012, 2013, 2014, 2015],
    "Sales": [250, 310, 410, 380, 470, 450]
}
df = pd.DataFrame(data)

# Create the line chart using plotly
fig = px.line(
    df,
    x="Year",
    y="Sales",
    title="Yearly Sales Growth",
    labels={"Sales": "Sales (in USD)", "Year": "Year"},
    markers=True
)

# Customize the layout (optional)
fig.update_layout(
    title_font_size=20,
    xaxis_title="Year",
    yaxis_title="Sales (in USD)",
    plot_bgcolor="rgba(240,240,240,0.9)"  # Light background
)

# Show the interactive chart
fig.show()
```

Bar Chart - matplotlib

```python
# Example data
products = ["Product A", "Product B", "Product C", "Product D"]
sales = [150, 200, 180, 220]

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='skyblue')

# Add chart elements
plt.title("Product Sales", fontsize=16)
plt.xlabel("Products")
plt.ylabel("Sales (in USD)")
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Add gridlines

# Show the chart
plt.show()
```

Bar Chart - seaborn

```python
# Example data
data = {"Product": ["Product A", "Product B", "Product C", "Product D"], 
        "Sales": [150, 200, 180, 220]}
df = pd.DataFrame(data)

# Create the bar chart using seaborn
sns.set(style="whitegrid")  # Optional: Set the style
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="Product", y="Sales", palette="Blues_d")

# Add chart elements
plt.title("Product Sales", fontsize=16)
plt.xlabel("Products")
plt.ylabel("Sales (in USD)")

# Show the chart
plt.show()
```

Bar Chart - plotly

```python
# Example data
data = {"Product": ["Product A", "Product B", "Product C", "Product D"], 
        "Sales": [150, 200, 180, 220]}
df = pd.DataFrame(data)

# Create the bar chart using plotly
fig = px.bar(
    df,
    x="Product",
    y="Sales",
    title="Product Sales",
    labels={"Product": "Products", "Sales": "Sales (in USD)"},
    color="Sales",  # Optional: Color bars by value
    text="Sales"    # Show sales values on bars
)

# Customize the layout (optional)
fig.update_layout(
    title_font_size=20,
    plot_bgcolor="rgba(240,240,240,0.9)"  # Light background
)

# Show the interactive chart
fig.show()
```

Histogram - matplotlib

```python
import matplotlib.pyplot as plt

# Example data: sales amounts (in USD)
sales = [120, 150, 100, 170, 130, 200, 220, 90, 110, 160, 140, 180]

# Create the histogram
plt.figure(figsize=(8, 5))
plt.hist(sales, bins=6, color='skyblue', edgecolor='black')

# Add chart elements
plt.title("Distribution of Sales Amounts", fontsize=16)
plt.xlabel("Sales Amount (in USD)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the chart
plt.show()
```

Histogram - seaborn

```python
# Example data: sales amounts (in USD)
sales = [120, 150, 100, 170, 130, 200, 220, 90, 110, 160, 140, 180]

# Create the histogram using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(sales, bins=6, kde=True, color="blue")

# Add chart elements
plt.title("Distribution of Sales Amounts", fontsize=16)
plt.xlabel("Sales Amount (in USD)")
plt.ylabel("Frequency")

# Show the chart
plt.show()
```

Histogram - plotly

```python
# Example data: sales amounts (in USD)
sales = [120, 150, 100, 170, 130, 200, 220, 90, 110, 160, 140, 180]
df = pd.DataFrame({"Sales": sales})

# Create the histogram using plotly
fig = px.histogram(
    df,
    x="Sales",
    nbins=6,
    title="Distribution of Sales Amounts",
    labels={"Sales": "Sales Amount (in USD)"},
    color_discrete_sequence=["blue"]
)

# Customize the layout
fig.update_layout(
    title_font_size=20,
    xaxis_title="Sales Amount (in USD)",
    yaxis_title="Frequency",
    plot_bgcolor="rgba(240,240,240,0.9)"
)

# Show the interactive chart
fig.show()
```

Pie Chart - matplotlib

```python
# Example data
products = ["Product A", "Product B", "Product C", "Product D"]
sales = [120, 150, 100, 170]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sales, labels=products, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green', 'red'])

# Add chart elements
plt.title("Sales Proportion by Product", fontsize=16)

# Show the chart
plt.show()
```

Pie Chart - seaborn

```python
# Example data
products = ["Product A", "Product B", "Product C", "Product D"]
sales = [120, 150, 100, 170]

# Create the pie chart using matplotlib, with seaborn style
sns.set(style="whitegrid")
plt.figure(figsize=(8, 8))
plt.pie(sales, labels=products, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Blues", len(products)))

# Add chart elements
plt.title("Sales Proportion by Product", fontsize=16)

# Show the chart
plt.show()
```

Pie Chart - plotly

```python
# Example data
data = {"Product": ["Product A", "Product B", "Product C", "Product D"], "Sales": [120, 150, 100, 170]}
df = pd.DataFrame(data)

# Create the pie chart using plotly
fig = px.pie(
    df,
    names="Product",
    values="Sales",
    title="Sales Proportion by Product",
    color="Product",
    color_discrete_sequence=["skyblue", "orange", "green", "red"]
)

# Show the interactive chart
fig.show()
```

Scatter Plot - matplotlib

```python
# Example data
products = ["Product A", "Product B", "Product C", "Product D"]
sales = [120, 150, 100, 170]
profits = [30, 50, 20, 60]

# Create the scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(sales, profits, color='blue', edgecolor='black', s=100)

# Add chart elements
for i, product in enumerate(products):
    plt.text(sales[i] + 2, profits[i] + 2, product, fontsize=12)

plt.title("Sales vs Profits", fontsize=16)
plt.xlabel("Sales (in USD)")
plt.ylabel("Profits (in USD)")
plt.grid(True)

# Show the chart
plt.show()
```

Scatter Plot - seaborn

```python
# Example data
data = {
    "Product": ["Product A", "Product B", "Product C", "Product D"],
    "Sales": [120, 150, 100, 170],
    "Profits": [30, 50, 20, 60]
}
df = pd.DataFrame(data)

# Create the scatter plot using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Sales", y="Profits", hue="Product", palette="Set1", s=100)

# Add chart elements
plt.title("Sales vs Profits", fontsize=16)
plt.xlabel("Sales (in USD)")
plt.ylabel("Profits (in USD)")

# Show the chart
plt.show()
```

Scatter Plot - plotly

```python
# Example data
data = {
    "Product": ["Product A", "Product B", "Product C", "Product D"],
    "Sales": [120, 150, 100, 170],
    "Profits": [30, 50, 20, 60]
}
df = pd.DataFrame(data)

# Create the scatter plot using plotly
fig = px.scatter(
    df,
    x="Sales",
    y="Profits",
    color="Product",
    title="Sales vs Profits",
    labels={"Sales": "Sales (in USD)", "Profits": "Profits (in USD)"},
    size="Sales",  # Optional: Size points by sales
    hover_name="Product"  # Show product name on hover
)

# Show the interactive chart
fig.show()
```
