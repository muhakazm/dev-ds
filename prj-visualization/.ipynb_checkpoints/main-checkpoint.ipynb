# Widely Used Chart Types for Data Visualization

## **Basic Charts**
1. **Line Chart**: To show trends over time or continuous data.
1. **Bar Chart**: To compare categories or groups.
1. **Histogram**: To show the distribution of a single variable.
1. **Pie Chart**: To show proportions within a whole.
1. **Scatter Plot**: To show relationships between two variables.

## **Advanced Charts**
1. **Box Plot**: To show the distribution and identify outliers.
1. **Heatmap**: To visualize data density or intensity across a two-dimensional space.
1. **Area Chart**: To represent cumulative values over time.
1. **Bubble Chart**: To compare three variables using position and size.
1. **Stacked Bar Chart**: To show part-to-whole relationships for multiple categories.

## **Specialized Charts**
1. **Radar Chart**: To compare multiple variables for a single category.
1. **Tree Map**: To visualize hierarchical data using nested rectangles.
1. **Waterfall Chart**: To show cumulative effects of sequential positive and negative values.
1. **Sunburst Chart**: To display hierarchical data in a circular form.
1. **Chord Diagram**: To show relationships or flows between different categories.

Radar Chart - matplotlib

```python
# Data: Sales performance metrics
metrics = ['Price Competitiveness', 'Customer Satisfaction', 'Quality', 'Availability', 'Marketing']
values = [7, 9, 8, 6, 7]  # Example scores out of 10
values += values[:1]  # Repeat the first value to close the chart

# Angle calculation
angles = np.linspace(0, 2 * np.pi, len(metrics), endpoint=False).tolist()
angles += angles[:1]  # Close the loop

# Plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2)

# Add labels
ax.set_yticks(range(1, 11, 2))  # Custom ticks
ax.set_xticks(angles[:-1])
ax.set_xticklabels(metrics)
ax.set_title('Product Sales Performance Radar Chart', y=1.1)

plt.show()
```

### Tree Maps

Tree Maps are particularly useful for exploring sales performance by hierarchical categories, such as regions and products, and for spotting areas that require strategic focus, such as low-margin products or underperforming regions.

Tree Map - squarify; matplotlib doesnt have a native function for treemaps, but you can use the squarify library as an alternative.

```python
# Data: Sales revenue and profit margin
categories = ['Electronics', 'Furniture', 'Groceries', 'Clothing', 'Toys']
sales = [50000, 30000, 20000, 15000, 10000]  # Revenue
profit_margin = [20, 15, 10, 5, 25]  # Profit margin (%)

# Normalize sales for squarify
norm_sales = [val / sum(sales) for val in sales]

# Define color based on profit margin
colors = plt.cm.viridis([pm / max(profit_margin) for pm in profit_margin])

# Plot
plt.figure(figsize=(8, 6))
squarify.plot(sizes=norm_sales, label=categories, color=colors, alpha=0.8)
plt.title('Sales Performance by Product Category')
plt.axis('off')  # Hide axes
plt.show()
```

Seaborn doesnt directly support Tree Maps either, but you can combine it with Matplotlib for styling.

### Waterfall Chart

A Waterfall Chart is used to visualize the cumulative effect of sequential positive and negative values. It is often used in financial or sales contexts, such as tracking changes in revenue, expenses, or profit over time.

Imagine you want to visualize how total revenue is affected by different factors such as product sales, discounts, returns, and shipping fees.

A Waterfall Chart for sales helps visualize:
1. Base Revenue: Starting point of total sales.
1. Individual Product Contributions: Identify which products contributed most to revenue.
1. Discounts and Returns: See how deductions affect overall revenue.
1. Final Total: Cumulative effect of all factors.

This type of chart is a valuable tool for financial analysts, product managers, and sales teams to make informed decisions about strategy and resource allocation.

Matplotlib does not have native support for Waterfall Charts, but you can build one manually.

```python
# Data: Revenue breakdown
categories = ['Base Revenue', 'Product A', 'Product B', 'Discounts', 'Returns', 'Shipping']
values = [50000, 20000, 15000, -5000, -7000, 2000]  # Positive/negative contributions

# Calculate cumulative sums
cumulative = np.cumsum([0] + values[:-1])

# Plot
plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values, color=['green' if v > 0 else 'red' for v in values], width=0.5)
plt.plot(np.arange(len(categories)), cumulative + values, color='blue', marker='o', label='Cumulative Total')
plt.axhline(0, color='black', linewidth=0.8)

# Annotate values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:+}', ha='center', va='bottom' if yval > 0 else 'top')

plt.title('Revenue Breakdown Waterfall Chart')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Seaborn does not natively support Waterfall Charts, but you can use its styling in combination with Matplotlib for a polished look.

### Sunburst Chart

A Sunburst Chart is a multi-level, radial visualization used to display hierarchical data. Its particularly useful for showing proportions across different levels of a dataset, such as sales by region, product category, and product type.

Imagine you want to analyze sales revenue broken down by Region, then Product Category, and finally Product Type.

Matplotlib does not have built-in support for Sunburst Charts. However, you can approximate a Sunburst Chart using mpl-sunburst or similar libraries, but the visualization will be more manual and less interactive.

Chord Diagram with Plotly (Sankey Alternative)

Although Plotly doesnt support Chord Diagrams directly, a Sankey Diagram can be used to visualize flows between groups.



```python
# import plotly.graph_objects as go

# Data: Customer flows between regions
labels = ["North", "South", "East", "West"]
source = [0, 0, 0, 1, 1, 2]  # From
target = [1, 2, 3, 2, 3, 3]  # To
values = [100, 50, 30, 40, 70, 20]  # Flow values

# Create Sankey Diagram
fig = go.Figure(go.Sankey(
    node=dict(
        pad=20,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels
    ),
    link=dict(
        source=source,
        target=target,
        value=values
    )
))

fig.update_layout(title_text="Customer Flows Between Regions", font_size=12)
fig.show()
```
