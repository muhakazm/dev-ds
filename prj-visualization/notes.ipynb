```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Checklist of things to keep in mind when making a plot

1. Understanding Your Data
    - Type of Data: Is it numerical, categorical, or time series?
    - Data Format: Ensure your data is clean and in the right format (e.g., no missing values, correct data types).
    - Distribution: Know if the data is skewed, has outliers, or follows a specific pattern.

1. Choosing the Right Plot
    - Histograms & KDE plots → For understanding distributions.
    - Scatter plots → For relationships between two variables.
    - Bar plots → For comparing categorical data.
    - Box plots → For spotting outliers and distribution.
    - Line plots → For trends over time.

1. Figure and Axis Configuration
    - Figure Size: Use plt.figure(figsize=(width, height)) to adjust based on dataset size and complexity.
    - Subplots: Use plt.subplot(rows, cols, index) to arrange multiple plots.
    - Aspect Ratio: Ensure your plots are not stretched or compressed.

1. Labels and Titles
    - Title (plt.title) → Give a clear, informative title.
    - Axis Labels (plt.xlabel, plt.ylabel) → Always label axes with units where applicable.
    - Legend (plt.legend()) → Helps in understanding multiple data series.

1. Aesthetics and Readability
    - Colors: Use distinguishable colors to avoid confusion.
    - Grid (plt.grid(True)) → Helps in better reading of values.
    - Tick Marks (plt.xticks, plt.yticks) → Adjust to avoid clutter.
    - Whitespace: Use plt.subplots_adjust() to prevent overlap in subplots.

1. Handling Datatypes
    - Numerical: Works well with histograms, scatter, and line plots.
    - Categorical: Use bar plots or box plots. Convert categorical data if necessary.
    - Time-Series: Ensure date formats are properly converted (pd.to_datetime()).

1. Avoid Common Mistakes
    - Check for NaN values: Use .dropna() or .fillna().
    - Avoid using sns.distplot() (Deprecated): Use sns.histplot() or sns.kdeplot().
    - Check scaling: Log-transform if necessary for better visualization (np.log1p(data)).

## `distplot()` Depreciation

Seaborn deprecated `sns.distplot()` because it was confusing—it combined histogram and KDE in a single function, which limited customization. Now, Seaborn encourages using `sns.histplot()` for histograms and `sns.kdeplot()` for KDE (kernel density estimation) separately.

```python
# import numpy as np
# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns

# Generate dummy data
np.random.seed(42)  # Ensure reproducibility
prices = np.random.exponential(scale=50, size=500)  # Skewed price distribution
df_usa = pd.DataFrame({'price': prices})  # Create DataFrame

# Visualization
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.title('Price Distribution', fontsize=14)

# Using histplot + kdeplot instead of distplot
sns.histplot(df_usa['price'], kde=True, bins=30, color='blue', alpha=0.6)
sns.kdeplot(df_usa['price'], color='red', linewidth=2)

# plt.xlabel('Price (US$)')
# plt.ylabel('Density')

plt.show()
```
