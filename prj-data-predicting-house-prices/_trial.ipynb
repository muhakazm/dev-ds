```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
import scipy.stats as st
import os
```

```python
df_usa = pd.read_csv(os.path.join('data','kc_house_data.csv'))
```

```python
df_usa.shape
```

```python
### check null values in price
df_usa['price'].isnull().any()
```

```python
df_usa[['price']].head(100)
```

```python
plt.figure(figsize=(12, 6))

### Distribution Plot (Histogram + KDE)
plt.subplot(121)
plt.title('Price Distribution', fontsize=14)
sns.histplot(df_usa['price'], kde=True, bins=30, color='blue', alpha=0.4)
sns.kdeplot(df_usa['price'], color='red', linewidth=2) ### what's this for?
plt.xlabel('Price (US$)')
plt.ylabel('Density')

plt.subplot(122)

plt.show()
```
